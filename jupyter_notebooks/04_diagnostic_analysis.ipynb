#### Benchmarking India Against Higher-Ranked Countries

This section compares India's scores with those of top-ranked countries to understand where India lags and what areas can be improved by learning from others.

### Steps:
- Select top 3 happiest countries from a given year
- Compare GDP, life expectancy, social support
- Visualize differences with bar plots or grouped comparisons


```python
# Select a specific year
year = 2019

# Get top 3 happiest countries based on Happiness Score
top_countries = df[df['Year'] == year].sort_values('Happiness Score', ascending=False).head(3)['Country'].tolist()

# Add India to the list
compare_countries = top_countries + ['India']

# Filter the data for these countries and the selected year
compare_data = df[(df['Year'] == year) & (df['Country'].isin(compare_countries))]

# Define relevant features using your dataset's column names
features = ['Economy', 'Healthy life expectancy', 'Family']

# Set the country as index for plotting
compare_data_plot = compare_data.set_index('Country')[features]

# Plot the comparison
compare_data_plot.plot(kind='bar', figsize=(10, 6))
plt.title(f"Benchmarking India vs Top Countries ({year})")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
